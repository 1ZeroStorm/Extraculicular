```python
class discountCal:

    def dc_student(self, price):
        return price * 0.8

    def dc_vip(self, price):
        return price * 0.7
    
    def dc_seasonal(self, price):
        return price * 0.9
    

if __name__ == "__main__":    
    price = 100
    dc = discountCal()
    print("Student price: ", dc.dc_student(price))
    print("VIP price: ", dc.dc_vip(price))
    print("Seasonal price: ", dc.dc_seasonal(price))
    

```

```python
class Engine:
    def start(self):
        print("Engine started")

class ElectricEngine(Engine):
    def charge(self):
        print("Charging battery")

class Car:
    def __init__(self):
        self.engine = Engine() 

class ElectricCar:
    def __init__(self):
        self.engine = ElectricEngine()
        

if __name__ == "__main__":
    car = Car()
    car.engine.start()
    electric_car = ElectricCar()
    electric_car.engine.start()
    electric_car.engine.charge()
```

```python
from abc import ABC, abstractmethod

class payment(ABC):
    @abstractmethod
    def credit_card_payment(self, amount):
        pass

    @abstractmethod
    def e_wallet_payment(self, amount):
        pass

class OnlinePayment(payment):
    def credit_card_payment(self, amount):
        print(f"Processing credit card payment of {amount}")

    def e_wallet_payment(self, amount):
        print(f"Processing e-wallet payment of {amount}")

if __name__ == "__main__":
    online_payment = OnlinePayment()
    online_payment.credit_card_payment(100)
    online_payment.e_wallet_payment(50)

# secara langsung bayar menggunakan metode yang diinginkan,
# tidak perlu membuat method pay maupun base class
```

```python
class StudentRepo:
    def __init__(self):
        self.__students = []

    def add_student(self, student):
        self.__students.append(student)
        
    def get_students(self):
        return self.__students
    
class StudentService:
    def __init__(self, student_repo_class):
        self.student_repo_class = student_repo_class

    def verify(self):
        for student in self.student_repo_class.get_students():
            print(f"Verifying student: {student.kelas} - {student.no_absen} - {student.nama} - Umur {student.umur}")

            if student.umur < 17:
                print("Student is underage")
            else:
                print("Student is of age")
        

class Student:
    def __init__(self, kelas, no_absen, umur, nama=None):
        self.kelas = kelas
        self.no_absen = no_absen
        self.umur = umur
        self.nama = nama

if __name__ == "__main__":
    student_repo = StudentRepo()
    student_service = StudentService(student_repo)
    student1 = Student("XII RPL 1", 1, 16, "John")
    student2 = Student("XII RPL 2", 2, 18, "Jane")
    student_repo.add_student(student1)
    student_repo.add_student(student2)
    student_service.verify()



```

```python
class Report:
    def __init__(self, repType):
        self.repType = repType

    def generate_report(self):
        self.repType.generate()
    
class type_pdf:
    def generate(self):
        print("Generating PDF report")

class type_excel:
    def generate(self):
        print("Generating Excel report")

class type_csv:
    def generate(self):
        print("Generating CSV report")

if __name__ == "__main__":

    for j in [type_pdf, type_excel, type_csv]:
        report = Report(repType=j())
        report.generate_report()
    
```
